```python
def AND(x1, x2):
    w1, w2, theta=0.5, 0.5, 0.7
    tmp=x1*w1 + x2*w2
    if tmp<=theta:
        return 0
    else:
        return 1
```

```python
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

```python
import numpy as np
x=np.array([0,1]) #입력
w=np.array([0.5, 0.5]) #가중치
b=-0.7 #편향
print(w*x)
print(np.sum(w*x))
print(np.sum(w*x)+b)
```

```python
def AND(x1,x2):
    x=np.array([x1, x2])
    w=np.array([0.5, 0.5]) #w1, w2는 각 입력 신호가 결과에 주는 영향을 조절하는 매개변수
    b=-0.7 # 뉴런이 얼마나 쉽게 활성화 하느냐를 조정하는 매개변수
    tmp=np.sum(w*x)+b
    if tmp<=0:
        return 0
    else:
        return 1
```

```python
def NAND(x1, x2):
    x=np.array([x1, x2])
    w=np.array([-0.5, -0.5])
    b=0.7
    tmp=np.sum(w*x)+b
    if tmp<=0:
        return 0
    else:
        return 1
```

```python
def OR(x1, x2):
    x=np.array([x1, x2])
    w=np.array([0.5, 0.5])
    b=-0.2
    tmp=np.sum(w*x)+b
    if tmp<=0:
        return 0
    else:
        return 1
```

```python
def XOR(x1, x2):
    s1=NAND(x1, x2)
    s2=OR(x1, x2)
    y=AND(s1, s2)
    return y
```

```python
print(XOR(0,0))
print(XOR(1,0))
print(XOR(0,1))
print(XOR(1,1))
```
