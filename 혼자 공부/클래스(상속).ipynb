# 1. 상속(inheritance) 이란?
클래스에서 상속이란, 물려주는 클래스(Parent Class)의 내용(속성, 메서드)을 물려받는 클래스(Child class, sub class)가 가지게 되는 것

```python
class Country:
    name='국가명'
    population='인구'
    capital='수도'

    def show(self):
        print('국가 클래스의 메소드입니다.')

class Korea(Country):
    def __init__(self, name):
        self.name=name
    
    def show_name(self):
        print('국가 이름은 : ', self.name)
```

```python
a=Korea('대한민국')
a.show()
```

```python
a.show_name()
```

```python
a.name
```

```python
a.capital
```

```python
class Korea(Country):
    def __init__(self, name, population, capital):
        self.name=name
        self.population=population
        self.capital=capital

    def show(self):
        print(
            """
            국가의 이름은 {} 입니다.
            국가의 인구는 {} 입니다.
            국가의 수도는 {} 입니다.
            """.format(self.name, self.population, self.capital)
        )
```

```python
class Korea(Country):
    def __init__(self, name, population, capital):
        self.name=name
        self.population=population
        self.capital=capital

    def show(self):
        super().show()
        print(
            """
            국가의 이름은 {} 입니다.
            국가의 인구는 {} 입니다.
            국가의 수도는 {} 입니다.
            """.format(self.name, self.population, self.capital)
        )
```

```python
a=Korea('대한민국',5000000, '서울')
a.show()
```

# 2. 정적메소드

```python
class CustomClass:
    def add_instance_method(self, a, b):
        return a+b
    
    #classmethod
    def add_class_method(cls,a,b):
        return a+b

    #staticmethod
    def add_static_method(a,b):
        return a+b
```

```python
CustomClass.add_instance_method(None, 3, 5)
```

```python
CustomClass.add_class_method(CustomClass, 3, 5)
```

```python
CustomClass.add_class_method(3, 5)
```

```python
CustomClass.add_static_method(3, 5)
```

```python
a=CustomClass()
a.add_class_method(3,5)
```

```python
a.add_static_method(3,5)
```

```python
class Language:
    default_language="English"

    def __init__(self):
        self.show='나의 언어는'+self.default_language

    def class_my_language(cls):
        return cls()

    def static_my_language():
        return Language()

    def print_language(self):
        print(self.show)

class KoreanLanguage(Language):
    default_language="한국어"
```

```python
a=KoreanLanguage.static_my_language()
b=KoreanLanguage.class_my_language()
```
