```python
#1-3 마겟과 머신러닝
import matplotlib.pyplot as plt

bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

#산점도는 x,y축으로 이뤄진 좌표계에 두 변수(x,y)의 관계를 표현하는 방법
plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
#일직선에 가까운 형태를 선형
```

```python
#빙어 데이터


smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

plt.scatter(smelt_length, smelt_weight)
plt.scatter(bream_length, bream_weight)
plt.xlabel("length")
plt.ylabel("weight")
plt.show()
```

```python
#K-최근접 아웃
#k-최근접 이웃 알고리즘을 구현한 클래스인 KNeighbors Classifier를 임포트
from sklearn.neighbors import KNeighborsClassifier

length=bream_length+smelt_length
weight=bream_weight+smelt_weight

fish_data=[[l,w] for l,w in zip(length, weight)]
fish_target = [1]*35 + [0]*14
print(fish_data)
kn=KNeighborsClassifier() #클래스의 객체 만들기
kn.fit(fish_data, fish_target)
'''
이 객체에 fish_data와 fish_target을 전달하여 도미를 찾기 위한 기준을 학습
이런 과정을 "훈련"이라고 함
사이킷런에서는 fit() 메서드가 이러한 역할을 함
'''
kn.score(fish_data, fish_target)
'''
이 객체가 얼마나 잘 훈련되었는지 평가
모델을 평가하는 메서드는 score()
이 메서드는 0과 1사이의 값을 반환
1은 모든 데이터를 정확히 맞췄다는 의미, 0.5는 절반을 맞췄다는 의미
'''
kn.predict([[30,600]])
```

```python
'''
마켓과 머신러닝
생선 분류 문제
도미 데이터 준비하기
'''
import matplotlib.pyplot as plt

bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

print("도미 그래프")
plt.scatter(bream_length, bream_weight)
plt.xlabel("length")
plt.ylabel("weight")
plt.show()
print()
#빙어 데이터 준비하기
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

print("빙어 그래프")
plt.scatter(smelt_length, smelt_weight)
plt.xlabel("length")
plt.ylabel("weight")
plt.show()
print()
#첫 번째 머신러닝 프로그램
length=bream_length+smelt_length
weight=bream_weight+smelt_weight

fish_data=[[l,w] for l,w in zip(length, weight)]
print(fish_data)
print()
fish_target=[1]*35 + [0]*14
print(fish_target)
print()

from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier()

kn.fit(fish_data, fish_target) #훈련
kn.score(fish_data, fish_target) #0과 1 사이로 적합한지 판단

#k-최근접 이웃 알고리즘
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
print("빙어+도미")
plt.scatter(30,600,marker='^')
plt.xlabel("length")
plt.ylabel("weight")
plt.show()

kn.predict([[30,600]])
print()
print(kn._fit_X) #fish 데이터
print()
print(kn._y) #타겟 데이터

kn49=KNeighborsClassifier(n_neighbors=49) #참고 데이터를 49개로 한 kn49모델
kn49.fit(fish_data, fish_target) #훈련
kn49.score(fish_data, fish_target) #예측

print(35/49)
```
