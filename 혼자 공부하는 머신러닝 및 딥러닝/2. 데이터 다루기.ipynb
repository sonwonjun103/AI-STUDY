```python
#2.2 데이터 전처리
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8, 
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7, 
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

import numpy as np
#넘파이의 column_stack() 함수는 전달받은 리스트를 일렬로 세운 다음 차례대로 나란히 연결 (튜를로 전달)
fish_data=np.column_stack((fish_length,fish_weight))
print(fish_data[:5])

#np.ones() 와 np.zeros()는 각각 원하는 개수의 1과 0을 채운 배열을 만들어줌
#np.concatenate()함수는 배열연결
fish_target=np.concatenate((np.ones(35), np.zeros(14)))#괄호 중요하네요
print(fish_target)

#사이킷런으로 훈련세트와 테스트 세트 나누기
#train_test_split은 전달되는 리스트나 배열을 비율에 맞게 훈련 세트와 테스트 세트로 나누어 준다
#이 함수는 기본적으로 25%를 테스트 세트로 떼어냄
#stratify 매개변수에 타깃 데이터를 전달하면 클래스 비율에 맞게 데이터를 나눔
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target=train_test_split(fish_data, fish_target, stratify=fish_target,random_state=42)

print(test_target)

from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)

print(kn.predict([[25, 150]]))

import matplotlib.pyplot as plt
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.xlabel("length")
plt.ylabel('weight')
plt.show()

print()
distances, indexes=kn.kneighbors([[25, 150]])
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 250, marker='^')
plt.scatter(train_input[indexes, 0],train_input[indexes, 1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

print()
print(distances)

#거리가 다른 이유는 x와 y의 축 기준이 다름
#바꾸기 위해 xlim, ylim 함수를 사용
print()
distances, indexes=kn.kneighbors([[25, 150]])
print("디스턴스", end=" ")
print(distances)
print("인덱스", end=" ")
print(indexes)
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 250, marker='^')
plt.scatter(train_input[indexes, 0],train_input[indexes, 1], marker='D')
plt.xlim((0,1000))
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

print()
#그림을 보면 두 특성(길이와 무게)의 값이 놓인 범위가 매우 다름.
#이를 두 특성의 스케일이 다르다고 함
#알고리즘들은 샘플 간의 거리에 영향을 많이 받으므로 제대로 사용하려면 특성값을 일정한 기준으로 맞춰 주어야 한다 -> 이러한 작업을 데이터 전처리라고 한다.
'''
가장 많이 사용하는 전처리 방법은 표준점수
표준점수는 각 특성값이 0에서 표준편차의 몇 배만큼 떨어져 있는지를 나타냄
이를 통해 실제 특성값의 크기와 상관없이 동일한 조건으로 비교 가능
'''
#특성마다 스케일이 다르므로 평균과 표준편차는 각 특성별로 계산해야 함, 이를 위해 axis=0
mean=np.mean(train_input, axis=0)#평균
std=np.std(train_input, axis=0)#표준편차

print(mean, std)
print()

#전처리 데이터로 모델 훈련하기
train_scaled=(train_input-mean)/std

plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(25, 150, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

print()
new=([25,150]-mean)/std
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0],new[1], marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

print()
#전처리를 사용해 다시 k-최근접 이웃 모델 실행
kn.fit(train_scaled, train_target)
test_scaled=(test_input-mean)/std
kn.score(test_scaled, test_target)

print()
print(kn.predict([new]))

#최종적으로 k-최근접 이웃을 구한 산점도
#특성을 표준점수로 바뀌었기 때문에 거리가 올바르게 나올것임
print()
distances, indexes=kn.kneighbors([new])
print(distances)
print(indexes)
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0], new[1], marker='^')
plt.scatter(train_scaled[indexes, 0],train_scaled[indexes, 1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```
